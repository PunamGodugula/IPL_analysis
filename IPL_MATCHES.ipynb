###Analysis of Null Values Distribution:

```python
# Filter columns with null values
columns_with_null = ipl_matches_df.columns[ipl_matches_df.isnull().any()]

# Calculate total null values and percentage null values for filtered columns
null_info_df = pd.DataFrame({
    'Total Null Values': ipl_matches_df[columns_with_null].isnull().sum(),
    '% of Null Values': np.round((ipl_matches_df[columns_with_null].isnull().mean()) * 100, 2)
})

# Apply background color gradient directly to the styled DataFrame
null_info_df.style.background_gradient(cmap='YlOrRd')
```

###Handling Null Values:
* To address this, we can employ strategies such as removing rows with null values or imputing them with appropriate measures.
* For instance, in numerical columns, we might consider replacing null values with the mean or median.

```python
#method column have the most null values so we are dropping the column
ipl_matches_df.drop(['method'],axis=1,inplace=True)
```

```python
ipl_matches_df['SuperOver'].fillna('NoResult',inplace=True)
ipl_matches_df['WinningTeam'].fillna('NoResult',inplace=True)
ipl_matches_df['Player_of_Match'].fillna('NoResult',inplace= True)
ipl_matches_df['Margin'].fillna(0,inplace = True)

ipl_matches_df.loc[(ipl_matches_df['Venue']== 'Dubai International Cricket Stadium'),['City']] = 'Dubai'
ipl_matches_df.loc[(ipl_matches_df['Venue'] == 'Sharjah Cricket Stadium'),['City']] = 'Sharjah'
```

###Correcting Data Types:
The format for "Date" column should be changed from object->DateTime

```python
ipl_matches_df['Date'] = pd.to_datetime(ipl_matches_df['Date'])
print(ipl_matches_df['Date'].dtype)
```

###Data Update: Column Values Transformation


```python
#unique team names
ipl_matches_df.Team1.unique()
```

```python
# Group by 'Team1', extract the years from 'Date', find the minimum and maximum years, and sort the result
min_years = ipl_matches_df.groupby('Team1')['Date'].min().dt.year.sort_values().reset_index()
max_years = ipl_matches_df.groupby('Team1')['Date'].max().dt.year.sort_values().reset_index()

# Merge the results into a single DataFrame
result = pd.merge(min_years, max_years, on='Team1', suffixes=('_min', '_max'))

result.style.background_gradient(cmap='YlOrRd')
```

Over the years few team names have been updated so we gonna change those old names to the new ones.
* eg: Delhi Daredevils -> Delhi Capitals

```python
# Define mapping of old team names to new ones
team_mapping = {
    'Kings XI Punjab': 'Punjab Kings',
    'Delhi Daredevils': 'Delhi Capitals',
    'Pune Warriors': 'Rising Pune Supergiant',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Gujarat Lions': 'Gujarat Titans',
    'Deccan Chargers': 'Sunrisers Hyderabad'
}

# Apply the mapping to the 'Team1', 'Team2', and 'WinningTeam' columns
ipl_matches_df[['Team1', 'Team2', 'WinningTeam','TossWinner']] = ipl_matches_df[['Team1', 'Team2', 'WinningTeam','TossWinner']].replace(team_mapping)
```

> **DISTRIBUTION OF WINNING TEAMS OVER SEASONS**




```python
# Grouping by Season and WinningTeam and calculating the size of each group
winning_teams_over_seasons = ipl_matches_df.groupby(['Season', 'WinningTeam']).size().unstack(fill_value=0)

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(winning_teams_over_seasons, cmap='viridis', annot=True, fmt='d', linewidths=.5)
plt.title('Distribution of Winning Teams Over Seasons')
plt.xlabel('Winning Team')
plt.ylabel('Season')
plt.xticks(rotation=30)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

>  **TEAMS AND THEIR WIN PERCENTAGE**

```python
# Concatenate Team1 and Team2 columns and count the occurrences of each team
total_matches = pd.concat([ipl_matches_df['Team1'], ipl_matches_df['Team2']]).value_counts().reset_index()

# Count the total wins for each team
total_matches['Total_wins'] = ipl_matches_df['WinningTeam'].value_counts().reset_index()['count']

# Rename columns
total_matches.rename(columns={'index': 'Teams', 'count': 'total_matches_played'}, inplace=True)

# Calculate win ratio
total_matches['win_ratio']= round((total_matches['Total_wins']/total_matches['total_matches_played']),2)*100

#styled background
total_matches.style.background_gradient(cmap='inferno')
```

OBSERVATIONS:
1. Mumbai Indians lead with the highest matches played (231) and wins (131), resulting in a 57% win ratio.
2. Sunrisers Hyderabad and Royal Challengers Bangalore follow closely, each with over 220 matches played and win ratios of 53% and 50% respectively.
4. Teams like Gujarat Titans and Lucknow Super Giants show promising win ratios (54% and 60%) despite fewer matches played.

 > ***FACTORS CONTRIBUTING TO TEAM WINS, SUCH AS TOSS DECISIONS OR VENUE***

> **1. TOSS WIN RATIO**

```python
# Count the number of toss wins for each team
toss_wins = ipl_matches_df['TossWinner'].value_counts().reset_index()

# Count the number of total wins for each team
total_wins = ipl_matches_df['WinningTeam'].value_counts().reset_index()

# Rename the columns for clarity
toss_wins.columns = ['Team', 'TossWins']
total_wins.columns = ['Team', 'TotalWins']

# Merge the toss_wins and total_wins DataFrames on the 'Team' column
combined_wins = pd.merge(total_wins,toss_wins, on='Team', how='outer')

# Fill NaN values with 0 (if any team has not won toss or match)
combined_wins.fillna(0, inplace=True)

# Calculate the chances of winning the game after winning the toss
combined_wins['TossWinRatio'] = combined_wins.apply(lambda row: (row['TotalWins'] / row['TossWins']) * 100 if row['TossWins'] != 0 else 0, axis=1)


# Adding toss win percentage column
combined_wins.style.background_gradient(cmap = 'inferno')
```

```python
combined_wins.plot(x='Team', y=['TossWins', 'TotalWins'], kind='bar', figsize=(12, 6))
plt.title('Toss Wins vs Total Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend(['Toss Wins', 'Total Wins'])
plt.tight_layout()
plt.show()
```

> **2. MAX WINING TEAM AT A VENUE**

```python
# Find the top 10 venues by count of matches
top_venues = ipl_matches_df['Venue'].value_counts().head(10).index

# Initialize a dictionary to store the winning count for each team at each venue
venue_team_win_count = {venue: ipl_matches_df[ipl_matches_df['Venue'] == venue]['WinningTeam'].value_counts().head(1) for venue in top_venues}

# Convert the dictionary into a DataFrame
venue_team_win_count_df = pd.DataFrame(venue_team_win_count).T

# Plotting the stacked bar plot
plt.figure(figsize=(12, 10))
venue_team_win_count_df.plot(kind='bar', stacked=True, cmap='tab10')
plt.title('Win Counts of Top 10 Venues by Venue and Team', fontsize=15, fontweight='bold')
plt.xlabel('Venue', fontsize=12)
plt.ylabel('Win Count', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Team', fontsize=10, bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

> **3. NUMBER OF WINS BY METHOD OF WINNING IN IPL**

```python
# Grouping by 'WonBy' and counting the number of wins for each winning method
win_counts = ipl_matches_df.groupby('WonBy')['WinningTeam'].count().reset_index()

# Renaming columns for clarity
win_counts.columns = ['WonBy', 'NumberOfWins']

# Sorting values by the number of wins
win_counts = win_counts.sort_values(by='NumberOfWins', ascending=False)

# Plotting the bar graph with seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='WonBy', y='NumberOfWins', data=win_counts, palette='viridis')
plt.title('Number of Wins by Method of Winning in IPL', fontsize=15, fontweight='bold')
plt.xlabel('Method of Winning', fontsize=12)
plt.ylabel('Number of Wins', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()

# Display the plot
plt.show()
```

> **1. TOTAL WINNING PERCENTAGE WHEN A TEAM WON THE TOSS**

```python
# Calculate the number of matches where the toss winner is the match winner
toss_winner_match_winner = (ipl_matches_df['TossWinner'] == ipl_matches_df['WinningTeam'])

# Count the occurrences of True and False values
toss_vs_matchwin_counts = toss_winner_match_winner.value_counts()

# Get labels and sizes for the pie chart
labels = ['Toss Winner Match Winner', 'Toss Winner Not Match Winner']
sizes = toss_vs_matchwin_counts / len(toss_winner_match_winner) * 100

# Define colors for the pie chart
colors = ['#4CAF50','#FF5733']  # Red and green colors

# Plot the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, pctdistance=0.85, textprops={'fontsize': 12, 'fontweight': 'bold'})
plt.title("Toss vs Match Win", fontweight="bold", fontsize=15)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the pie chart
plt.show()
```

> **2. TOSS DESCISION WHEN A TEAM WON A TOSS**

```python
# Grouping by 'TossDecision' and counting the number of toss decisions for each decision
toss_decision_counts = ipl_matches_df.groupby('TossDecision')['TossWinner'].count()

# Plotting the pie chart
plt.figure(figsize=(4,3))
plt.pie(toss_decision_counts, labels=toss_decision_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen'])
plt.title('Toss Decisions Distribution', fontsize=15, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
